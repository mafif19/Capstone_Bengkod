# **3. Pemodelan dan Training**

```python
import time
import psutil
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import all_estimators
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

### **Balancing Data**

```python
# Definisikan kolom
target_column = 'NObeyesdad'
X = df.drop(columns=target_column)
y = df[target_column]

# Split 80:20
X_train_base, X_test, y_train_base, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

Pada bagian ini kamin memisahkan fitur (X) dan target (y), lalu membagi dataset menjadi data latih (80%) dan data uji (20%) menggunakan stratifikasi berdasarkan label target agar distribusi kelas tetap seimbang pada kedua subset. Parameter random_state=42 memastikan hasil pembagian konsisten setiap kali dijalankan.

```python
# Metode sampling yang digunakan
sampling_methods = {
    "No Sampling": None,
    "Random Oversampling": RandomOverSampler(random_state=42),
    "Random Undersampling": RandomUnderSampler(random_state=42),
    "SMOTE": SMOTE(random_state=42)
}
```

Kami mendefinisikan empat metode sampling yang akan digunakan dalam eksperimen, yaitu No Sampling, Random Oversampling, Random Undersampling, dan SMOTE. Metode-metode ini membantu kami menangani ketidakseimbangan kelas dalam data latih, sehingga model dapat belajar dari distribusi data yang lebih merata dan adil.

### **Algoritma**

```python
# Algoritma yang akan digunakan
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

Kami mengimpor beragam algoritma klasifikasi dari pustaka scikit-learn dan XGBoost yang akan digunakan dalam proses pelatihan model. Algoritma yang kami pilih meliputi pendekatan linear, pohon keputusan, ensemble, metode berbasis jarak, hingga jaringan saraf. Tujuannya adalah untuk menguji performa berbagai jenis model pada data yang telah disampling.

### **Models**

```python
models = {
    "LogReg": LogisticRegression(max_iter=1000),
    "KNN": KNeighborsClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "GradientBoosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    "NaiveBayes": GaussianNB(),
    "SVM": SVC(),
    "MLP": MLPClassifier(max_iter=1000)
}
```

Kami mendefinisikan sembilan model klasifikasi dalam sebuah dictionary bernama models. Masing-masing model mewakili pendekatan yang berbeda, seperti regresi logistik, KNN, pohon keputusan, ensemble, SVM, dan jaringan saraf (MLP). Parameter tambahan seperti max_iter dan eval_metric disesuaikan untuk memastikan proses pelatihan berjalan optimal.

### **Proses Training**

```python
results = []

for sm_name, sampler in sampling_methods.items():
    print(f"\n📦 Sampling: {sm_name}")

    # Apply sampling
    if sampler:
        X_train, y_train = sampler.fit_resample(X_train_base, y_train_base)
    else:
        X_train, y_train = X_train_base.copy(), y_train_base.copy()

    for model_name, model in models.items():
        start_cpu = psutil.cpu_percent(interval=None)
        start_time = time.time()

        try:
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)

            acc = accuracy_score(y_test, y_pred)
            prec = precision_score(y_test, y_pred, average='macro', zero_division=0)
            rec = recall_score(y_test, y_pred, average='macro', zero_division=0)
            f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)
            end_time = time.time()
            cpu_used = psutil.cpu_percent(interval=None) - start_cpu

            results.append({
                "Sampling": sm_name,
                "Model": model_name,
                "Accuracy": acc,
                "Precision": prec,
                "Recall": rec,
                "F1": f1,
                "Time (s)": end_time - start_time,
                "CPU (%)": cpu_used
            })
        except Exception as e:
            print(f"❌ Model {model_name} failed: {e}")
```

Kami menjalankan eksperimen klasifikasi dengan menguji semua kombinasi metode sampling dan algoritma pembelajaran. Untuk setiap metode sampling, data latih disesuaikan terlebih dahulu (jika diperlukan), kemudian masing-masing model dilatih dan diuji menggunakan data uji. Kami mencatat metrik evaluasi penting seperti akurasi, presisi, recall, F1-score, waktu komputasi, dan penggunaan CPU. Semua hasil disimpan dalam list results untuk analisis dan perbandingan selanjutnya. Proses ini memastikan penilaian yang adil terhadap performa tiap model pada kondisi distribusi data yang berbeda.

```python
# Hasil Pelatihan
results_df = pd.DataFrame(results)
top5 = results_df.sort_values(by="Accuracy", ascending=False).head(5)

print("\n🎯 Top 5 Models berdasarkan akurasi:")
display(top5)

print("\n📊 Tabel komparasi model:")
display(results_df.sort_values(by="Accuracy", ascending=False).reset_index(drop=True))
```

```python
# Set style
sns.set(style="whitegrid")

# Plotting Top 5 Models based on Accuracy
plt.figure(figsize=(12, 6))
top5_sorted = results_df.sort_values(by="Accuracy", ascending=True)
sns.barplot(data=results_df, x="Accuracy", y="Model", hue="Sampling", palette="viridis")
plt.title("Hasil Training Model")
plt.xlabel("Akurasi")
plt.ylabel("Model")
plt.legend(title="Sampling")
plt.tight_layout()
plt.show()

# Barplot Waktu Pelatihan
plt.figure(figsize=(14, 6))
sns.barplot(
    x="Time (s)", y="Model", hue="Sampling",
    data=results_df.sort_values(by="Time (s)", ascending=False)
)
plt.title("Waktu Pelatihan Model per Metode Sampling")
plt.xlabel("Waktu (detik)")
plt.ylabel("Model")
plt.legend(title="Sampling")
plt.tight_layout()
plt.show()

# Barplot Penggunaan CPU
plt.figure(figsize=(14, 6))
sns.barplot(
    x="CPU (%)", y="Model", hue="Sampling",
    data=results_df.sort_values(by="CPU (%)", ascending=False)
)
plt.title("Penggunaan CPU Model per Metode Sampling")
plt.xlabel("CPU (%)")
plt.ylabel("Model")
plt.legend(title="Sampling")
plt.tight_layout()
plt.show()
```

Kami menggabungkan semua hasil evaluasi ke dalam sebuah DataFrame, lalu menyortirnya berdasarkan nilai akurasi secara menurun untuk menampilkan lima model terbaik. Hasilnya menunjukkan bahwa kombinasi metode random oversampling dengan model RandomForest memberikan akurasi tertinggi sebesar 96.15%, diikuti oleh model RandomForest tanpa sampling, serta XGBoost dengan random oversampling.

Selain akurasi, kami juga membandingkan metrik lain seperti presisi, recall, F1-score, waktu komputasi, dan penggunaan CPU. Data ini memberikan gambaran menyeluruh terhadap efektivitas dan efisiensi tiap model, sehingga dapat membantu dalam memilih model yang optimal tidak hanya dari sisi performa, tetapi juga dari sisi sumber daya yang digunakan.

### **Evaluasi Model Terbaik**

```python
from sklearn.metrics import confusion_matrix, classification_report

# Evaluasi model terbaik
best = top5.iloc[0]
best_sampling = best['Sampling']
best_model_name = best['Model']

# Terapkan ulang sampling terbaik
if sampling_methods[best_sampling]:
    X_train_best, y_train_best = sampling_methods[best_sampling].fit_resample(X_train_base, y_train_base)
else:
    X_train_best, y_train_best = X_train_base.copy(), y_train_base.copy()

# Train model terbaik
best_model = models[best_model_name]
best_model.fit(X_train_best, y_train_best)
y_pred_best = best_model.predict(X_test)

# Jika pakai label angka saja
class_labels = sorted(np.unique(y_test))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_best)
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title(f"Confusion Matrix for Best Model ({best_model_name} with {best_sampling})")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

# Classification Report
print("\n📝 Classification Report:")
print(classification_report(y_test, y_pred_best, target_names=[str(c) for c in class_labels]))
```

Berdasarkan hasil pelatihan dan evaluasi, kombinasi metode SMOTE dengan algoritma Random Forest menghasilkan performa klasifikasi terbaik. Confusion matrix menunjukkan prediksi yang sangat presisi pada hampir seluruh kelas, dengan jumlah kesalahan klasifikasi yang sangat rendah. Evaluasi menggunakan classification report memberikan metrik berikut:

**Akurasi keseluruhan: 96%**

Rata-rata makro (macro average) untuk:

* Presisi: 0.96

* Recall: 0.95

* F1-score: 0.95

Kinerja model sangat baik terutama pada kelas-kelas yang memiliki jumlah data yang seimbang maupun tidak seimbang, menunjukkan bahwa penerapan SMOTE berhasil meningkatkan representasi kelas minoritas tanpa mengorbankan akurasi kelas mayoritas.
