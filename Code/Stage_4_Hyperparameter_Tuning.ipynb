```python
import time
import psutil
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.utils import all_estimators
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import GridSearchCV, ParameterGrid
```

Library yang Kami Gunakan:

- time & tqdm: Kami menggunakan `time` untuk mengukur durasi pelatihan setiap model, serta `tqdm` untuk memberikan progress bar yang memudahkan dalam memantau proses iteratif, khususnya saat tuning hyperparameter yang memakan waktu.

- psutil: Kami menggunakan library ini untuk memantau penggunaan CPU dan memori selama proses pelatihan model. Ini penting agar sistem tetap stabil saat menjalankan eksperimen yang intensif secara komputasi.

- sklearn.model_selection: Kami memanfaatkan `train_test_split` untuk membagi data secara acak menjadi data latih dan data uji, serta `GridSearchCV` dan `ParameterGrid` untuk mencari kombinasi hyperparameter terbaik yang dapat meningkatkan performa model kami.

- sklearn.metrics: Digunakan untuk mengevaluasi kinerja model berdasarkan metrik utama seperti akurasi, presisi, recall, dan F1-score, yang kami anggap penting dalam menilai kualitas klasifikasi.

- sklearn.utils: Dengan `all_estimators`, kami dapat secara dinamis mengakses berbagai algoritma klasifikasi dari scikit-learn, sehingga memungkinkan eksplorasi model yang lebih luas.

- imblearn.over_sampling & under_sampling: Kami menggunakan teknik sampling seperti `RandomOverSampler`, `SMOTE`, dan `RandomUnderSampler` untuk menangani ketidakseimbangan kelas dalam data. Ini kami lakukan agar model tidak bias terhadap kelas mayoritas, yang dapat menurunkan akurasi pada kelas minoritas.


### **Balancing Data**

```python
# Definisikan kolom
target_column = 'NObeyesdad'
X = df.drop(columns=target_column)
y = df[target_column]

# Split 80:20
X_train_base, X_test, y_train_base, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

Pada bagian ini, kami memisahkan fitur prediktor dan target dari dataset yang digunakan:

- target_column = 'NObeyesdad': Kolom NObeyesdad merupakan label atau target yang ingin kami prediksi, yaitu kategori obesitas seseorang berdasarkan sejumlah variabel gaya hidup dan kesehatan.

- X = df.drop(columns=target_column): Kami menyimpan semua kolom selain NObeyesdad ke dalam variabel X, yang berisi fitur-fitur yang akan digunakan untuk prediksi.

- y = df[target_column]: Variabel y menyimpan nilai target yaitu label kelas obesitas.

Selanjutnya, kami membagi data menjadi data latih dan data uji:

- Fungsi train_test_split() dari scikit-learn digunakan untuk membagi data secara acak tetapi terkontrol, dengan:

- test_size=0.2: Menunjukkan bahwa 20% dari total data digunakan sebagai data uji, sementara sisanya (80%) digunakan untuk melatih model.

- random_state=42: Digunakan agar proses pembagian data bisa direproduksi (reproducible)  hasilnya akan sama setiap kali dijalankan.

```python
# Metode sampling yang digunakan
sampling_methods = {
    "No Sampling": None,
    "Random Oversampling": RandomOverSampler(random_state=42),
    "Random Undersampling": RandomUnderSampler(random_state=42),
    "SMOTE": SMOTE(random_state=42)
}
```

Kami mendefinisikan empat metode sampling yang akan digunakan dalam eksperimen, yaitu No Sampling, Random Oversampling, Random Undersampling, dan SMOTE. Metode-metode ini membantu kami menangani ketidakseimbangan kelas dalam data latih, sehingga model dapat belajar dari distribusi data yang lebih merata dan adil.
Walaupun pada pelatihan sebelumnya hasil akurasi terbaik didapatkan ketika menggunakan metode sampilng random oversampling.

### **Algoritma**

```python
# Algoritma yang akan digunakan
from sklearn.ensemble import RandomForestClassifier
```

Dalam eksperimen klasifikasi ini, kami menggunakan algoritma Random Forest Classifier, yang merupakan salah satu metode ensemble learning berbasis pohon keputusan (decision tree). Random Forest bekerja dengan membangun banyak pohon keputusan selama proses pelatihan, dan menghasilkan prediksi akhir berdasarkan mayoritas suara (voting) dari semua pohon. Algoritma ini kami gunakan karena  pada pelatihan sebelumnya algoritma ini yang menghasilkan akurasi tertinggi hingga mencapai **96.15%**

### **Parameter yang akan diuji**

```python
# Parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
```

Untuk mendapatkan performa terbaik dari algoritma Random Forest, kami melakukan hyperparameter tuning dengan menyusun grid parameter. Grid ini berisi berbagai kombinasi nilai untuk hyperparameter penting dari model Random Forest.

Penjelasan Masing-masing Parameter:
- **n_estimators**: Jumlah pohon dalam hutan. Semakin banyak pohon, semakin stabil model, namun waktu pelatihan juga meningkat.

- **max_depth**: Kedalaman maksimum setiap pohon. Jika None, pohon akan tumbuh sampai semua daun murni. Membatasi kedalaman bisa mengurangi overfitting.

- **min_samples_split**: Jumlah minimal sampel yang dibutuhkan untuk membagi node. Nilai lebih tinggi akan menghasilkan pohon yang lebih sederhana.

- **min_samples_leaf**: Jumlah minimum sampel yang harus ada di daun pohon. Meningkatkan nilai ini bisa membantu mencegah overfitting.

- **bootstrap**: Menentukan apakah bootstrap samples digunakan saat membangun pohon. True artinya menggunakan sampling dengan pengembalian.

Tujuan:
Grid ini nantinya akan digunakan dalam proses GridSearchCV atau eksplorasi sistematis untuk menemukan kombinasi parameter terbaik berdasarkan skor akurasi atau metrik evaluasi lainnya.

Dengan pendekatan ini, kami berharap dapat mengoptimalkan performa klasifikasi serta meningkatkan generalisasi model terhadap data yang belum pernah dilihat sebelumnya.

### **Proses Training**

```python
param_combinations = list(ParameterGrid(param_grid))
results = []

# Loop kombinasi sampling dan hyperparameter
for sm_name, sampler in sampling_methods.items():
    for params in tqdm(param_combinations, desc=f"Sampling: {sm_name}"):
        if sampler:
            X_train, y_train = sampler.fit_resample(X_train_base, y_train_base)
        else:
            X_train, y_train = X_train_base.copy(), y_train_base.copy()

        model = RandomForestClassifier(random_state=42, **params)

        start_cpu = psutil.cpu_percent(interval=None)
        start_time = time.time()

        try:
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)

            acc = accuracy_score(y_test, y_pred)
            prec = precision_score(y_test, y_pred, average='macro', zero_division=0)
            rec = recall_score(y_test, y_pred, average='macro', zero_division=0)
            f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)
            end_time = time.time()
            cpu_used = psutil.cpu_percent(interval=None) - start_cpu

            results.append({
                "Sampling": sm_name,
                "Params": params,
                "Accuracy": acc,
                "Precision": prec,
                "Recall": rec,
                "F1": f1,
                "Time (s)": end_time - start_time,
                "CPU (%)": cpu_used
            })

        except Exception as e:
            continue
```

Dalam tahap ini, kami melakukan eksperimen komprehensif untuk mengevaluasi performa Random Forest dengan berbagai kombinasi metode sampling dan hyperparameter. Proses ini bertujuan untuk mencari kombinasi terbaik yang menghasilkan performa optimal dalam klasifikasi kategori obesitas.

Pertama, kami menghasilkan semua kemungkinan kombinasi parameter berdasarkan param_grid yang telah ditentukan sebelumnya menggunakan ParameterGrid dari sklearn.

Kami melakukan iterasi bersarang: pertama berdasarkan metode sampling (No Sampling, Random Oversampling, Random Undersampling, dan SMOTE), kemudian untuk setiap kombinasi parameter.

Di dalamnya:

- Data latih (X_train_base dan y_train_base) di-resample sesuai metode sampling yang dipilih.

- Model RandomForestClassifier dibuat menggunakan kombinasi hyperparameter saat ini.

- Kami mencatat waktu pelatihan dan penggunaan CPU dengan bantuan modul time dan psutil.

Setelah model dilatih dan diuji pada data uji (X_test), kami mengevaluasi kinerjanya menggunakan metrik:

- Akurasi, untuk melihat rasio prediksi benar.

- Presisi dan Recall makro, untuk mengukur performa di seluruh kelas secara seimbang.

- F1 Score, sebagai harmonisasi presisi dan recall.

- Hasil dari setiap kombinasi kemudian disimpan dalam list results.

Proses ini dilakukan total 4 kali (untuk setiap metode sampling), dengan 216 kombinasi parameter di setiap metode. Berikut waktu rata-rata untuk menyelesaikan tiap metode:

- Tanpa Sampling: ~2 menit 21 detik

- Random Oversampling: ~2 menit 26 detik

- Random Undersampling: ~1 menit 42 detik

- SMOTE: ~2 menit 54 detik

```python
# Hasil Pelatihan
results_df = pd.DataFrame(results)
results_df_sorted = results_df.sort_values(by="Accuracy", ascending=False).reset_index(drop=True)
top10 = results_df_sorted.head(10)


# Tampilkan seluruh isi dictionary Params tanpa dipotong
pd.set_option("display.max_colwidth", None)

# Konversi Params ke string agar tampil penuh
results_df["Params"] = results_df["Params"].apply(lambda x: str(x))

# Urutkan berdasarkan akurasi
results_df_sorted = results_df.sort_values(by="Accuracy", ascending=False).reset_index(drop=True)

# Ambil 10 model terbaik
top10 = results_df_sorted.head(10)

# Tampilkan
print("\nTop 10 Models berdasarkan akurasi:")
display(top10)
```

Pada tahap ini, kami menampilkan 10 kombinasi model terbaik dari seluruh eksperimen hyperparameter tuning yang telah dilakukan. Evaluasi dilakukan terhadap berbagai kombinasi parameter Random Forest, serta variasi metode sampling untuk menangani ketidakseimbangan data. Kami menyimpan seluruh metrik hasil pelatihan ke dalam sebuah DataFrame dan mengurutkannya berdasarkan nilai akurasi secara menurun.

Sebelumnya, untuk memastikan informasi konfigurasi hyperparameter (Params) tampil secara lengkap tanpa terpotong, kami mengatur ulang pengaturan tampilan Pandas dengan pd.set_option("display.max_colwidth", None), dan mengubah kolom dictionary tersebut menjadi bentuk string agar seluruh informasi dapat terlihat utuh dan mudah dianalisis.

Dari hasil yang ditampilkan, seluruh model dengan akurasi tertinggi (sekitar 97.01%) menggunakan metode SMOTE (Synthetic Minority Oversampling Technique). Ini menunjukkan bahwa SMOTE secara konsisten memberikan performa terbaik dalam mengatasi ketidakseimbangan data pada dataset ini.

Beberapa hal yang dapat disimpulkan dari hasil ini:

- Model dengan akurasi tertinggi menggunakan kombinasi bootstrap=True, max_depth=30, min_samples_split=5, min_samples_leaf=1, dan n_estimators=200.

- Nilai metrik lainnya seperti Precision, Recall, dan F1-Score juga menunjukkan performa yang seimbang dan tinggi, dengan nilai rata-rata di atas 0.96, yang mengindikasikan model bekerja secara konsisten di semua kelas.

- Waktu pelatihan (Time (s)) bervariasi dari sekitar 0.3 hingga 1 detik per model, yang masih tergolong efisien.

- Persentase penggunaan CPU (CPU (%)) menunjukkan nilai negatif karena cara pengukuran psutil.cpu_percent() yang tidak terlalu akurat jika dijalankan cepat secara berurutanini bisa diabaikan dalam konteks ini atau diperbaiki dengan pendekatan pengukuran lain jika diperlukan.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Buat label model singkat untuk ditampilkan di Y axis
def make_label(row):
    p = eval(row['Params']) if isinstance(row['Params'], str) else row['Params']
    return f"n={p['n_estimators']}, depth={p['max_depth']}, split={p['min_samples_split']}, leaf={p['min_samples_leaf']}, boot={p['bootstrap']}"

top10['Model_Label'] = top10.apply(make_label, axis=1)

# === PLOT 1: Accuracy ===
plt.figure(figsize=(14, 7))
sns.barplot(
    x="Accuracy",
    y="Model_Label",
    data=top10,
    hue="Sampling",
    dodge=False,
    palette="Set2"
)
plt.title("Top 10 Model Combinations by Accuracy", fontsize=14)
plt.xlabel("Accuracy", fontsize=12)
plt.ylabel("Model Hyperparameter Combination", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Anotasi nilai akurasi di ujung batang
for i, (acc, label) in enumerate(zip(top10["Accuracy"], top10["Model_Label"])):
    plt.text(acc + 0.001, i, f"{acc:.4f}", va='center')

plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

# === PLOT 2: Training Time ===
plt.figure(figsize=(14, 7))
sns.barplot(
    x="Time (s)",
    y="Model_Label",
    data=top10,
    hue="Sampling",
    dodge=False,
    palette="Set3"
)
plt.title("Top 10 Model Combinations by Training Time", fontsize=14)
plt.xlabel("Training Time (s)", fontsize=12)
plt.ylabel("Model Hyperparameter Combination", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Anotasi nilai waktu di ujung batang
for i, (t, label) in enumerate(zip(top10["Time (s)"], top10["Model_Label"])):
    plt.text(t + 0.01, i, f"{t:.2f}s", va='center')

plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

Visualisasi ini menyajikan 10 kombinasi model terbaik berdasarkan metrik akurasi dan waktu pelatihan. Setiap model diwakili oleh kombinasi parameter seperti n_estimators, max_depth, dan metode sampling yang digunakan. Kami menambahkan label hiperparameter langsung di grafik untuk mempermudah interpretasi dan evaluasi antar model.

- Grafik pertama menunjukkan bahwa sebagian besar model terbaik berasal dari kombinasi SMOTE dan parameter seperti n_estimators = 200 serta max_depth = 30.

- Grafik kedua menunjukkan perbandingan waktu pelatihan masing-masing model, yang penting untuk mempertimbangkan efisiensi.

### **Evaluasi Model Terbaik**

```python
from sklearn.metrics import confusion_matrix, classification_report

# Evaluasi model terbaik
best = top10.iloc[0]
best_sampling = best['Sampling']
best_params = best['Params']

if sampling_methods[best_sampling]:
    X_train_best, y_train_best = sampling_methods[best_sampling].fit_resample(X_train_base, y_train_base)
else:
    X_train_best, y_train_best = X_train_base.copy(), y_train_base.copy()

best_model = RandomForestClassifier(random_state=42, **best_params)
best_model.fit(X_train_best, y_train_best)
y_pred_best = best_model.predict(X_test)

class_labels = sorted(np.unique(y_test))

cm = confusion_matrix(y_test, y_pred_best)
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title(f"Confusion Matrix for Best Model (RandomForest + {best_sampling})")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

report = classification_report(y_test, y_pred_best, target_names=[str(c) for c in class_labels], output_dict=True)
report_df = pd.DataFrame(report).transpose()

results_df_sorted.head(1), report_df
```

Setelah melakukan proses pencarian kombinasi terbaik antara metode sampling dan hyperparameter Random Forest, kami memilih model dengan akurasi tertinggi untuk evaluasi lebih mendalam. Model terbaik ini menggunakan metode sampling SMOTE dan parameter hyperparameter yang optimal berdasarkan hasil tuning. Data latih kemudian disesuaikan kembali menggunakan metode sampling tersebut sebelum model dilatih ulang. Evaluasi dilakukan menggunakan data uji yang tidak pernah dilihat sebelumnya dengan menghasilkan confusion matrix dan classification report. Hasil evaluasi menunjukkan performa model yang sangat baik dengan akurasi mencapai sekitar 97%, presisi, recall, dan F1-score makro yang juga berada pada kisaran 96-97%. Confusion matrix memperlihatkan model mampu memprediksi sebagian besar kelas dengan tepat, sementara classification report mengonfirmasi keseimbangan antara presisi dan recall di berbagai kelas. Hal ini mengindikasikan bahwa kombinasi metode sampling SMOTE dan pengaturan hyperparameter Random Forest yang kami gunakan berhasil mengatasi masalah ketidakseimbangan data sekaligus menghasilkan model yang akurat dan andal.
